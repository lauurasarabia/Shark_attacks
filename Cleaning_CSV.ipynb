After cleaning, we split duplicated data saved in one row:

```python
attacks = attacks.assign(Age=attacks.Age.str.split('&')).explode('Age')
attacks
```

As we can see above, we find not exact data and they can affect to our analysis.
For instance: "18 to 22" is two different age group "Teenager" or "Adult".
That's why we extract only digits:

```python
attacks.Age = attacks.Age.str.extract('(\d+)')
attacks.Age
```

```python
# Remove "NaN" values
attacks.dropna(subset=['Age'], inplace=True)
```

```python
attacks["Age"].isna().sum()
```

```python
#Change type to int in order to be able to work with age ranges
attacks['Age'] = attacks['Age'].astype(int)
```

```python
attacks["Age"].dtypes
```

```python
# Create age ranges by conditions:
conditions = [
    (attacks['Age'] >= 0) & (attacks['Age'] <= 12),
    (attacks['Age'] >= 13) & (attacks['Age'] <= 18),
    (attacks['Age'] >= 19) & (attacks['Age'] <= 64),
    (attacks['Age'] >= 65) & (attacks['Age'] <= 100),
    ]

values = ["Child", "Teenager", "Adult", "Senior"]

attacks["Age Group"] = np.select(conditions, values)
attacks
```

SPECIES

```python
# Remove NaN values
attacks.dropna(subset=['Species'], inplace=True)
```

```python
# We transform the "Species" to lowercase in order to be able to filter
attacks['Species'] = attacks['Species'].str.lower().str.strip()
```

```python
attacks["Species"].value_counts().head(10)
```

```python
# Main shark species: 
# White Shark, Tiger Shark, Hammerhead shark, Bull Shark, Galapagos Shark, Nurse Shark. 
pattern = r'(white shark|tiger shark|bull shark|hammerhead shark|galapagos shark|nurse shark)'
    
attacks['Top Species'] = attacks['Species'].str.extract(pattern, expand=True)
attacks.drop(columns='Species', inplace=True)   
    
```

```python
attacks.dropna(subset=['Top Species'], inplace=True)
```

```python
attacks["Top Species"].value_counts()
```

INJURY

```python
# 1 NaN value in "Injury", we remove it:
attacks.dropna(subset=['Injury'], inplace=True)
```

```python
# We transform the "Injury" to lowercase in order to be able to filter
attacks["Injury"] = attacks["Injury"].str.lower()
```

Create a ranking of points by part of the body üèÜ

```python
# ARMS 10 POINTS
arm_search = r'(arm|hand|finger|elbow|forearm|wrist)'

attacks['Arm Pts'] = attacks['Injury'].str.extract(arm_search, expand = True)
attacks['Arm Pts'].fillna("", inplace=True)
attacks.loc[attacks['Arm Pts'] != "", 'Arm Pts'] = 10
attacks['Arm Pts'] = pd.to_numeric(attacks['Arm Pts'],errors='coerce')
```

```python
# LEGS 20 POINTS
leg_search = r'(leg|ankle|groin|thigh|knee|foot|feet|buttocks)'

attacks['Leg Pts'] = attacks['Injury'].str.extract(leg_search, expand = True)
attacks['Leg Pts'].fillna("", inplace=True)
attacks.loc[attacks['Leg Pts'] != "", 'Leg Pts'] = 20
attacks['Leg Pts'] = pd.to_numeric(attacks['Leg Pts'],errors='coerce')
```

```python
# BODY 50 POINTS
body_search = r'(body|ribs|heart|back|waist|hip|belly|abdomen|chest)'

attacks['Body Pts'] = attacks['Injury'].str.extract(body_search, expand = True)
attacks['Body Pts'].fillna("", inplace=True)
attacks.loc[attacks['Body Pts'] != "", 'Body Pts'] = 50
attacks['Body Pts'] = pd.to_numeric(attacks['Body Pts'],errors='coerce')
```

```python
# HEAD 100 POINTS
head_search = r'(head|skull|throat|face|eyes|ear|neck|cheeks|nose|forehead)'

attacks['Head Pts'] = attacks['Injury'].str.extract(head_search, expand = True)
attacks['Head Pts'].fillna("", inplace=True)
attacks.loc[attacks['Head Pts'] != "", 'Head Pts'] = 100
attacks['Head Pts'] = pd.to_numeric(attacks['Head Pts'],errors='coerce')
```

```python
# Delete the original column as we will work with the new ones:
attacks.drop(columns='Injury', inplace=True)
attacks
```

```python
attacks.reset_index(drop=True, inplace=True)
```

```python
attacks['Total Pts'] = attacks[['Arm Pts', 'Leg Pts', 'Body Pts', 'Head Pts']].sum(axis=1)
```

```python
attacks['Arm Pts'] = attacks['Arm Pts'].fillna(0)
attacks['Leg Pts'] = attacks['Leg Pts'].fillna(0)
attacks['Body Pts'] = attacks['Body Pts'].fillna(0)
attacks['Head Pts'] = attacks['Head Pts'].fillna(0)
```
