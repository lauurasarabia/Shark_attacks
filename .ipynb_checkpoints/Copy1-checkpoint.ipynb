```python
arm_search = r'(arm|hand|finger|elbow|forearm|wrist)'
leg_search = r'(leg|ankle|groin|thigh|knee|foot|feet|buttocks)'
body_search = r'(body|ribs|heart|back|waist|hips|belly|abdomen|chest)'
head_search = r'(head|skull|throat|face|eyes|ear|neck|cheeks|nose|forehead)'
```

```python
attacks["Arm Injuries"] = attacks['Injury'].str.extract(arm_search, expand = True)
attacks["Leg Injuries"] = attacks['Injury'].str.extract(leg_search, expand = True)
attacks["Body Injuries"] = attacks['Injury'].str.extract(body_search, expand = True)
attacks["Head Injuries"] = attacks['Injury'].str.extract(head_search, expand = True)
```
